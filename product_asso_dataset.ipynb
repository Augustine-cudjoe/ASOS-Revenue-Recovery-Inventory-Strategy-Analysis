Data Loading 

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

df=pd.read_csv('products_asos.csv')
df.info()
```

Data cleaning

```python
df['price']=pd.to_numeric(df['price'], errors='coerce')
df=df.dropna(subset='price')

print(f" Data loaded :{len(df) } rows - {df.columns} ")
```

```python
df['description']=df['description'].astype(str)

def get_brand(text):
    if 'by' in text:
        try:
            return text.split('by')[1].split( ' ')[0]
        except:
            return 'unknown'
        return 'unknown'
    
df['brand_raw']=df['description'].apply(get_brand)
df
```

```python
brand_map={
    'New':'New Look',
    'River':'River Island',
    'Miss':'Miss Selfridge',
    'TopShowWelcome':'Topshop'
}

df['brand']=df['brand_raw'].map(brand_map).fillna('brand_raw')

brand_count=df['brand'].value_counts()
valid_brand= brand_count[brand_count > 5].index
df_clean=df[df['Brand'].isin(valid_brand)].copy()

df_clean['Brand'].value_counts().head(5)
```
